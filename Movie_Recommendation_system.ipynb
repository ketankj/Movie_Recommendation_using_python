```python
# importing files
import numpy as np
import pandas as pd
# if user makes mistake giving movie name then it will give closest match of the value
import difflib
# convert textual data into numerical values
from sklearn.feature_extraction.text import TfidfVectorizer
# find which movies have heighest similarity score
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# collecting data and preprocess
# loading the data
movies_data = pd.read_csv("movies.csv")
```

```python
# printing first 5 rows
movies_data.head()
```

```python
# finding number of rows and columns
movies_data.shape
```

```python
# selecting relevant features for recommendation

selected_features = ['genres','keywords','tagline','cast','director','overview','original_language']
print(selected_features)
```

```python
# replacing the missing values with null string

for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the 8 selected features

combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']+' '+movies_data['overview']+' '+movies_data['original_language']
```

```python
print(combined_features)
```

```python
# converting text data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# using cosine similarity to get similarity score
#it will find similar values ie. similar movies

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
# getting the movie name from the user
movie_name = input("Enter your Favourate Movie Name!")
```

```python
# creating list of names of all the movies.
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
# finding the close match for the movie name given by the user.
# taking the value given by the user and comparing it with all the values in the list.
find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
# finding the index of the movie with the title
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# getting list of similar movies based by index of the movie
# to loop through the list we are using enumarate.
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sort the movies based on higher similarity score
# reverse = true means descending order
#x:x[1] means x is similarity score and x[0] is the first element and second element is similarity value

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
#print top 10 names of similar movies based on index

print("Top 10 movies suggested for you!")
i = 1
for movie in sorted_similar_movies:
    index = movie[0]
    # using the index value finding the title from the movies dataset
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    if(i<11):
        print(i,'.',title_from_index)
        i+=1
```
